```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x=pd.read_csv('trainx.csv')
```

```python
y=pd.read_csv('trainy.csv')
```

```python
model=tf.keras.Sequential([tf.keras.layers.Dense(30,input_shape=(27,),activation='tanh',name='Dense1'),
                           tf.keras.layers.Dense(30,activation='tanh',name='Dense2'),
                           #tf.keras.layers.Dense(30,activation='tanh',name='Dense3'),
                           tf.keras.layers.Dense(1,name='Dense3')])
```

```python
model.compile(optimizer='adam',
              loss='mse')
model.fit(x,y,epochs=1000)
testx=pd.read_csv('testx.csv')
testy=pd.read_csv('testy.csv')
error=(model.predict(testx)-testy)
print(error)
np.savetxt('output.txt',error)
```

```python
weight1,bias1 = model.get_layer('Dense1').get_weights()
np.savetxt('weight1.txt',weight1)
np.savetxt('bias1.txt',bias1)
```

```python
weight3,bias3 = model.get_layer('Dense3').get_weights()
np.savetxt('weight3.txt',weight3)
np.savetxt('bias3.txt',bias3)
```

```python
weight2,bias2 = model.get_layer('Dense2').get_weights()
np.savetxt('weight2.txt',weight2)
np.savetxt('bias2.txt',bias2)
```
